```python
with fits.open('dvt.fits', mode="readonly") as hdulist:
    
    # Extract stellar parameters from the primary header.  We'll get the effective temperature, surface gravity,
    # and TESS magnitude.
    star_teff = hdulist[0].header['TEFF']
    star_logg = hdulist[0].header['LOGG']
    
    # Extract some of the fit parameters for the first TCE.  These are stored in the FITS header of the first
    # extension.
    period = hdulist[1].header['TPERIOD']
    duration = hdulist[1].header['TDUR']
    epoch = hdulist[1].header['TEPOCH']
    depth = hdulist[1].header['TDEPTH']
    
    # Extract some of the columns of interest for the first TCE signal.  These are stored in the binary FITS table
    # in the first extension.  We'll extract the timestamps in TBJD, phase, initial fluxes, and corresponding
    # model fluxes.
    times = hdulist[1].data['TIME']
    phases = hdulist[1].data['PHASE']
    fluxes_init = hdulist[1].data['LC_INIT']
    model_fluxes_init = hdulist[1].data['MODEL_INIT']
```

```python
# First sort the phase and flux arrays by phase so we can draw the connecting lines between points.
sort_indexes = np.argsort(phases)

# Start figure and axis.
fig, ax = plt.subplots(figsize=(12,4))

# Plot the detrended fluxes as black circles.  We will plot them in sorted order.
ax.plot(phases[sort_indexes], fluxes_init[sort_indexes], 'ko',
       markersize=2)

# Plot the model fluxes as a red line.  We will plot them in sorted order so the line connects between points cleanly.
ax.plot(phases[sort_indexes], model_fluxes_init[sort_indexes], '-r')

# Label the axes for the figure.
ax.set_ylabel("Flux (relative)")
ax.set_xlabel("Orbital Phase")



plt.show()
```

```python
#Plot the first image of the FLUX Column with WCS overlay.
tpf_data = hdul[1].data
first_image = tpf_data['FLUX'][1]

#The aperture extension header contains the same WCS as that in the Pixels extension.
wcs = WCS(hdul[2].header)

#Plot the first image in the file.
fig = plt.figure(figsize = (6,6))
fig.add_subplot(111, projection = wcs)
plt.imshow(first_image, origin = 'lower', cmap = plt.cm.viridis)
plt.xlabel('RA', fontsize = 14)
plt.ylabel('Dec', fontsize = 14)
plt.grid(axis = 'both', color = 'white', ls = 'solid')
```

```python
ap_image = (hdul[2].data)
ap_want = np.bitwise_and(ap_image, 2) / float(2)
print(ap_want)
```

```python
def aperture_phot(image,aperture):
    
    flux = np.sum(image[aperture == 1])

    return flux
```

```python
#Use the map lambda functions to apply that function to all cadences
opap_flux = np.array(list (map (lambda x: aperture_phot(x,ap_want), tpf_data['FLUX'] ) ) )
print(opap_flux[1:10])
```

```python
time_bjd = tpf_data["TIME"]
tpf_head = hdul[1].header
bjd_ref = tpf_head['BJDREFI'] + tpf_head['BJDREFF']
time = time_bjd + bjd_ref
print("TESS JD Offset is: %f" % bjd_ref)
```

```python
plt.figure(figsize = (13,4))
plt.plot(time_bjd, opap_flux,'.', label = 'Optimal Aperture', ms = 6)
plt.legend(loc = "lower left")
plt.xlabel('Time (BJD)', fontsize = 14)
plt.ylabel('Flux (e-/s)', fontsize = 14)
plt.ylim(43000,44500)
plt.title("Kepler Lightcurve Derived from Calibrated Pixels.", fontsize = 14)
```

```python
import os
from astropy.timeseries import BoxLeastSquares as bls
from sklearn.impute import SimpleImputer
from astropy import units as u
```

```python
def open_tpf(path):
 
    
    hdul = fits.open(path)
    
    time_nom = hdul[1].data['TIME']
    
    flux = hdul[1].data['FLUX']
    
    return {'time_nom': time_nom, 'flux' : flux}, hdul

#opening the first 4 tpf_files.
data1, hdul1 = open_tpf('./test (1).fits') 
data2, hdul2 = open_tpf('./test (2).fits')
data3, hdul3 = open_tpf('./test (3).fits')
data4, hdul4 = open_tpf('./test (4).fits')
```

```python
def remove_nan(flux):
    
    """Function to remove the nan values from the flux 
       and replace it with the mean values of other values 
       in the flux array."""
    
    f = np.asarray(flux)
    
    f_corr = np.nan_to_num(f)
    
    mean = np.mean(f_corr[f_corr > 0])
    
    f_corr[f_corr == 0] = mean
    
    return f_corr

#removing the nan values from the flux data of first 4 tpf files.
flux1_corr = remove_nan(data1['flux'])
flux2_corr = remove_nan(data2['flux'])
flux3_corr = remove_nan(data3['flux'])
flux4_corr = remove_nan(data4['flux'])
```

```python
#Plotting image obtained from test (4).fits  file during its exposure.

wcs4 = WCS(hdul4[2].header)
fig = plt.figure(figsize=(6, 6))
fig.add_subplot(111, projection = wcs4)
fig.set_size_inches((10, 8))
plt.imshow(flux4_corr[0], origin = 'lower', cmap = plt.cm.viridis)
plt.xlabel('RA', fontsize = 14)
plt.ylabel('Dec', fontsize = 14)
plt.grid(axis = 'both', color = 'white', ls = 'solid')
plt.colorbar()

```

```python
def get_time(tpf_head, time_nom):
    
    """adding the time offset to the nominal time to
       get the real time values."""
    
    bjd_ref = tpf_head['BJDREFI'] + tpf_head['BJDREFF']
    
    time_real = time_nom + bjd_ref
    
    return ({'start': time_real[0], 'end': time_real[len(time_real)-1]}, time_real)


t_interval1, t_series1 = get_time(hdul1[1].header, data1['time_nom'])
t_interval2, t_series2 = get_time(hdul2[1].header, data2['time_nom'])
t_interval3, t_series3 = get_time(hdul3[1].header, data3['time_nom'])
t_interval4, t_series4 = get_time(hdul4[1].header, data4['time_nom'])
```

```python
def find_aper(aperature):
    """Function to determine The aperature positions 
       at which the pixel values should be added."""
    
    ap_image = aperature
    
    ap_req = np.bitwise_and(ap_image, 2) / float(2)
    
    return ap_req



def aperture_phot(image, aperture):
    """adding the pixel values at the optimal aperture 
       positions."""
    
    flux = np.sum(image[aperture == 1])

    return flux



def get_flux(flux_arr, aperature):
    """Function which returns the optimal flux using the 
       above 2 helper functions."""
    
    ap_req = find_aper(aperature)
    
    op_ap_flux = np.array(list (map (lambda x: aperture_phot(x,ap_req), flux_arr ) ) )
    
    return op_ap_flux
```

```python
#Optimal flux for the first 4 test.fits files.

opt_flux1 = get_flux(flux1_corr, hdul1[2].data)
opt_flux2 = get_flux(flux2_corr, hdul2[2].data)
opt_flux3 = get_flux(flux3_corr, hdul3[2].data)
opt_flux4 = get_flux(flux4_corr, hdul4[2].data)
```

```python
#Combining the optimal flux data and the time series data for the first 4 tpf(test.fits) files.

opt_flux = np.concatenate((opt_flux4, opt_flux3, opt_flux2, opt_flux1), axis=0)
t_series = np.concatenate((t_series4, t_series3, t_series2, t_series1), axis=0)
```

```python
#Optimal light curve for the test(4).fits file
plt.figure(figsize = (10, 8))
plt.plot(t_series4, opt_flux4,'.', label = 'Optimal Aperture', ms = 4)
plt.legend(loc = "lower left")
plt.xlabel('Time (REAL)', fontsize = 20)
plt.ylabel('Flux (e-/s)', fontsize = 20)
plt.ylim(39000, 41000)
plt.title("Lightcurve Derived from Flux.", fontsize = 20)
```

```python
#Replacing the outliers in the opt_flux data with the mean value. 
m = np.mean(opt_flux4[opt_flux4 > 39000])

opt_flux4[opt_flux4 < 39000] = m



#Using the BoxLeastSquares algorithm on the test(4).fits file and 
#Plotting the log likelihood curve.
model = bls(data4['time_nom']*u.day, opt_flux4, hdul4[1].data['TIMECORR'])
periodogram = model.autopower(0.1, objective='likelihood')
plt.plot(periodogram.period, periodogram.power)
pl = plt.gcf()
pl.set_size_inches(10, 8)
plt.show()
```

```python
#Using the period value where log likelihood is maximum to get other relevant 
#Information such as transit_duration, transit_time and transit_period. 
max_power = np.argmax(periodogram.power)
duration = periodogram.duration[max_power]
transit_period = periodogram.period[max_power]
transit_time = periodogram.transit_time[max_power]
print('duration:', duration)
print('period:', transit_period)
print('transit_time:', transit_time)
```

```python
#Using relevant data obtained from log likelihood curve we try to find a 
#step function which best fits the lightcurve for test(4).fits .
fitted_curve = model.model(data4['time_nom'], transit_period, duration, transit_time)

print(np.unique(fitted_curve))
```

```python
#Superimposing the lightcurve for test(4).fits and a step plot
#obtained in the fiited_curve.
plt.figure(figsize = (10, 8))
plt.plot(data4['time_nom'], opt_flux4,'.', label = 'Optimal Aperture', ms = 4)
plt.step(data4['time_nom'], fitted_curve)
plt.legend(loc = "lower left")
plt.xlabel('Time (days)', fontsize = 20)
plt.ylabel('Flux (e-/s)', fontsize = 20)
plt.ylim(39000, 41000)
plt.title("Lightcurve Derived from Flux.", fontsize = 20)
```
